```python
train = pd.read_csv('train.csv')
depths = pd.read_csv('depths.csv')
train = train.merge(depths,how='inner',on='id')
```

```python
def expand(df_x):
    d = {'id':[],'rle_mask':[],'z':[]}
    for j in range(df_x.shape[0]):
        for i in range(4):
            d['id'].append(df_x.iloc[j]['id']+'_0'+str(i))
            d['rle_mask'].append(df_x.iloc[j]['rle_mask'])
            d['z'].append(df_x.iloc[j]['z'])
    return pd.DataFrame(d)
```

```python
print(train.shape)
train.head()
```

```python
df_y = expand(train)
```
